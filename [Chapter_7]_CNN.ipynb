앞 장 까지는 완전 연결 계층인 Affine 계층과 활성화 함수로 신경망을 구성했다면, 합성곱에서는 합성곱 계층-활성화 함수(ReLU)-풀링 계층으로 구성된다.(풀링 계층은 생략하기도 한다.) 
> Affine-ReLU 에서, Conv-ReLU-(Pooling)으로 바뀜

단, 출력에 가까운 층에서는 Affine-ReLU 구성을 사용할 수 있고, 마지막 출력 계층에서는 Affine-softmax 조합을 그대로 사용한다. 이는 일반적인 CNN에서 흔히 볼 수 있는 구성이다.

* CNN 특징
  * 패딩(padding)
    * **합성곱 연산을 수행하기 전에** 입력 데이터 주변을 특정 값으로 채우는 것
    * 출력 크기를 조절할 목적으로 쓰인다.
  * 스트라이드(stride)
    * 필터를 적용하는 위치의 간격
  * 각 계층 사이에는 3차원 데이터와 같이 입체적인 데이터가 흐른다는 점에서 완전 연결 신경망과는 다르다.
