```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
import os
import glob
```

```python
# 데이터 로드
os.chdir('/Users/yun-yeongjun/Desktop/Technique100/PythonML-main/PythonML-main/chapter01')

m_store = pd.read_csv('m_store.csv')
m_area = pd.read_csv('m_area.csv')

order_all = pd.DataFrame()
tbl_order_file = os.path.join(os.getcwd(), 'tbl_order_*.csv')
tbl_order_files = glob.glob(tbl_order_file)

for file in tbl_order_files:
    order_data = pd.read_csv(file)
    print(f'{file}:{len(order_data)}')
    order_all = pd.concat([order_all, order_data], ignore_index=True)
    
order_all
```

#### 6. 불필요한 데이터를 제거하자

```python
# m_store를 다시 보면 store_id에 보수담당으로 999가 있음 -> 불필요
# store_id = 999는 유지보수용 데이터로, order 테이블에도 있을 것으로 생각된다.
m_store
```

```python
order_all
```

```python
order_all = order_all.loc[order_all['store_id'] != 999]
order_all
```

* store_id = 999인 데이터를 제거하니 데이터가 707706개에서 703886개로 줄어들었다.
    * 이런 유지보수용 데이터는 언제 만들어졌는지 알 수 없기 때문에 미리 제거하는 것이 좋다.
   
***
* loc: location의 약어
    * df의 행이나 컬럼에 label이나 boolean array로 접근
    * 인간이 읽을 수 있는 label값으로 데이터에 접근한다.
    * 헷갈리는 메서드로 iloc가 있음
* iloc: integer location의 약어
    * df의 행이나 컬럼에 index값으로 접근
    * 컴퓨터가 읽을 수 있는 indexing값으로 데이터에 접근한다.

```python
# 데이터가 줄어들었으니 통계량에도 변화가 생겼을 것이다.
order_all.describe()
```

#### 마스터 데이터를 조인하자.

```python
# 마스터 데이터를 가로로 조인
# 매장별 주문 건수를 확인하기 위해 m_store테이블을 order_all에 조인
# 각 데이터의 공통열은 'store_id'
# how= 로 주축이 될 테이블을 정한다.

order_data = pd.merge(order_all, m_store, on='store_id', how='left')
order_data
```

```python
# order_data에 지역 마스터를 조인

order_data = pd.merge(order_data, m_area, on='area_cd', how='left')
order_data
```

#### 마스터가 존재하지 않는 코드에 이름을 설정하자

* order_data에서 takeout_flag, status 컬럼에 값은 있지만, 이 값이 무엇을 의미하는지 테이블만 봐서는 알 수 없다.
    * 책에서는 takeout_flag 컬럼의 값 중, 0= delivery, 1= takeout이라고 한다.
    * 또한 status 컬럼의 값 중, 0: 주문 접수, 1: 지불 완료, 2: 배달 완료, 9: 주문 취소

```python
# order_data에 takeout_name 컬럼을 추가하고 각각의 값에 연결

order_data.loc[order_data['takeout_flag'] == 0, 'takeout_name'] = 'delivery'
order_data.loc[order_data['takeout_flag'] == 1, 'takeout_name'] = 'takeout'
order_data
```

```python
# status도 위처럼 추가

order_data.loc[order_data['status'] == 0, 'status_name'] = '주문 접수'
order_data.loc[order_data['status'] == 1, 'status_name'] = '지불 완료'
order_data.loc[order_data['status'] == 2, 'status_name'] = '배달 완료'
order_data.loc[order_data['status'] == 9, 'status_name'] = '주문 취소'

order_data
```

#### 분석 기초 테이블을 파일에 저장하자.

```python
# 나는 os.chdir로 경로를 애초에 설정해두거나 따로 경로를 저장한 변수로 바로 to_csv 저장하는 것을 선호
# 책에서는(24p) 경로 상에 폴더를 생성하는 os.makedir 메서드를 사용하여 저장하는 방식, 이게 더 편한 것 같기도..

store_path = '/Users/yun-yeongjun/Desktop/Technique100/make_data'
store_file = os.path.join(store_path, 'order_data.csv')
order_data.to_csv(store_file, index=False)
```
